## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
def rotate_matrix(matrix, direction):
    if direction == 'left':
        return [list(row) for row in zip(*matrix)][::-1]
    if direction == 'right':
        return [list(row)[::-1] for row in zip(*matrix)]
    return "Please enter 'left' or 'right'"


def main():
    data = input().split()
    direction = data[-1].lower()
    nums = list(map(int, data[:-1]))

    n = int(len(nums) ** 0.5)
    if n * n != len(nums):
        print("Size error")
        return

    matrix = [nums[i:i+n] for i in range(0, len(nums), n)]

    result = rotate_matrix(matrix, direction)

    if isinstance(result, str):
        print(result)
    else:
        for row in result:
            print(*row)


main()
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
class Book:
    def __init__(self, title, author, year, isbn, number_of_copies):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.number_of_copies = number_of_copies
        print(f"Book {title} created")

    def get_info(self):
        return f"<{self.title}> {self.author}, {self.year}, {self.isbn}, copies: {self.number_of_copies}"

    def change_number(self, delta):
        if self.number_of_copies + delta < 0:
            return False
        self.number_of_copies += delta
        return True


class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.list_books = []
        print(f"User {name} created")

    def add_book_to_list(self, book):
        if book in self.list_books:
            return False
        self.list_books.append(book)
        return True

    def remove_book_from_list(self, book):
        if book not in self.list_books:
            return False
        self.list_books.remove(book)
        return True

    def list_borrowed_books(self):
        return [b.get_info() for b in self.list_books]


class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.transaction_history = []
        print("Library created")

    def add_book(self, book):
        if self.find_book_by_isbn(book.isbn) is not None:
            return False
        self.books.append(book)
        return True

    def delete_book_by_isbn(self, isbn):
        book = self.find_book_by_isbn(isbn)
        if book is None:
            return False
        for u in self.users:
            for b in u.list_books:
                if b.isbn == isbn:
                    return False
        self.books.remove(book)
        return True

    def register_user(self, user):
        if self.find_user_by_card(user.card_number) is not None:
            return False
        self.users.append(user)
        return True

    def delete_user_by_card(self, card_number):
        user = self.find_user_by_card(card_number)
        if user is None:
            return False
        if user.list_books:
            return False
        self.users.remove(user)
        return True

    def issue_books(self, card_number, isbn):
        user = self.find_user_by_card(card_number)
        book = self.find_book_by_isbn(isbn)
        if user is None or book is None:
            return False
        if not book.change_number(-1):
            return False
        if not user.add_book_to_list(book):
            book.change_number(+1)
            return False
        self.transaction_history.append(f"{user.name} issued {book.title}")
        return True

    def return_books(self, card_number, isbn):
        user = self.find_user_by_card(card_number)
        book = self.find_book_by_isbn(isbn)
        if user is None or book is None:
            return False
        if not user.remove_book_from_list(book):
            return False
        book.change_number(+1)
        self.transaction_history.append(f"{user.name} returned {book.title}")
        return True

    def search_book(self, title=None, author=None, year=None, isbn=None):
        result = []
        for b in self.books:
            ok = True
            if title is not None and title.lower() not in b.title.lower():
                ok = False
            if author is not None and author.lower() not in b.author.lower():
                ok = False
            if year is not None and year != b.year:
                ok = False
            if isbn is not None and isbn != b.isbn:
                ok = False
            if ok:
                result.append(b.get_info())
        return result

    def search_user(self, name=None, card_number=None):
        result = []
        for u in self.users:
            ok = True
            if name is not None and name.lower() not in u.name.lower():
                ok = False
            if card_number is not None and card_number != u.card_number:
                ok = False
            if ok:
                result.append(f"{u.name}, card: {u.card_number}, borrowed: {len(u.list_books)}")
        return result

    def display_transaction(self):
        for record in self.transaction_history:
            print(record)

    def find_book_by_isbn(self, isbn):
        for b in self.books:
            if b.isbn == isbn:
                return b
        return None

    def find_user_by_card(self, card_number):
        for u in self.users:
            if u.card_number == card_number:
                return u
        return None


````

## Problem 6*

Let
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python
def forward(n):
    s = 0.0
    for i in range(1, n + 1):
        s += 1.0 / (i * i)
    return s

def reverse(n):
    s = 0.0
    for i in range(n, 0, -1):
        s += 1.0 / (i * i)
    return s

A = forward(10000)
B = reverse(10000)

print("A =", format(A, ".20f"))
print("B =", format(B, ".20f"))
print("A - B =", format(A - B, ".20e"))
````
